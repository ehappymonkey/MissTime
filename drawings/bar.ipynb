```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# =========================
# Real data
# =========================
missing_ratios = np.array([25, 50, 75])
backbones = ["TimesNet", "TimeMixer"]

vanilla = {
    "TimesNet":  np.array([0.303, 0.537, 0.773]),
    "TimeMixer": np.array([0.256, 0.524, 0.757]),
}

misstime = {
    "TimesNet":  np.array([0.130, 0.175, 0.245]),
    "TimeMixer": np.array([0.116, 0.147, 0.203]),
}

# =========================
# Style
# =========================
plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["STIXGeneral", "DejaVu Serif"],
    "mathtext.fontset": "stix",
    "font.size": 18,
    "axes.labelsize": 19,
    "xtick.labelsize": 16,
    "ytick.labelsize": 17,
    "legend.fontsize": 12.8,
    "axes.linewidth": 1.0,
    "pdf.fonttype": 42,
    "ps.fonttype": 42,
})

fig, ax = plt.subplots(figsize=(6.8, 4.5))

palette = {
    "TimeMixer": {
        "vanilla": "#E3E3E3",
        "misstime": "#BDBDBD",
        "edge": "#7A7A7A",
    },
    "TimesNet": {
        "vanilla": "#DDF3DF",
        "misstime": "#8CCF8C",
        "edge": "#5E9E5F",
    },
}

bar_width = 0.25
ratio_gap = 0.78
backbone_gap = 0.82

x_centers = {}
current_x = 0.0

for backbone in backbones:
    centers = current_x + np.arange(len(missing_ratios)) * ratio_gap
    x_centers[backbone] = centers
    current_x = centers[-1] + ratio_gap + backbone_gap

all_values = np.concatenate([
    np.concatenate(list(vanilla.values())),
    np.concatenate(list(misstime.values())),
])

ymax = all_values.max() * 1.15
ymin = 0.0

legend_handles = []

for backbone in backbones:
    centers = x_centers[backbone]
    v = vanilla[backbone]
    m = misstime[backbone]
    colors = palette[backbone]

    ax.bar(
        centers - bar_width / 2,
        v,
        width=bar_width,
        color=colors["vanilla"],
        edgecolor="black",
        linewidth=0.25,
        alpha=0.96,
        label=backbone,
        zorder=3,
    )

    ax.bar(
        centers + bar_width / 2,
        m,
        width=bar_width,
        color=colors["misstime"],
        edgecolor="black",
        linewidth=0.25,
        label=f"{backbone}+MissTime",
        zorder=3,
    )

    legend_handles.extend([
        Patch(
            facecolor=colors["vanilla"],
            edgecolor="none",
            label=backbone,
        ),
        Patch(
            facecolor=colors["misstime"],
            edgecolor="none",
            label=f"{backbone}+MissTime",
        ),
    ])

    # MSE reduction labels
    for i in range(len(centers)):
        reduction = (v[i] - m[i]) / v[i] * 100
        ax.text(
            centers[i] + bar_width / 2,
            m[i] + ymax * 0.018,
            f"↓{reduction:.1f}%",
            ha="center",
            va="bottom",
            fontsize=12.2,
            color="black",
            fontweight="normal",
        )

# =========================
# X-axis labels
# =========================
all_ratio_positions = np.concatenate([x_centers[b] for b in backbones])
all_ratio_labels = [f"{r}%" for _ in backbones for r in missing_ratios]

ax.set_xticks(all_ratio_positions)
ax.set_xticklabels(all_ratio_labels)

# Vertical separator between backbones
for i in range(len(backbones) - 1):
    right_edge = x_centers[backbones[i]][-1]
    left_edge = x_centers[backbones[i + 1]][0]
    sep = (right_edge + left_edge) / 2
    ax.axvline(
        sep,
        color="#D3D3D3",
        linestyle="--",
        linewidth=1.0,
        zorder=0,
    )

ax.set_xlabel("Missing Ratio (%)", labelpad=10)
ax.set_ylabel("MSE", labelpad=12)

ax.set_ylim(ymin, ymax)

ax.grid(
    axis="y",
    linestyle="--",
    linewidth=0.9,
    color="#D8D8D8",
    alpha=0.70,
    zorder=0,
)

for spine in ["top", "bottom", "left", "right"]:
    ax.spines[spine].set_linewidth(1.0)

ax.tick_params(axis="both", width=1.3, length=5, direction="out")

ax.legend(
    handles=legend_handles,
    loc="upper center",
    bbox_to_anchor=(0.5, 0.99),
    frameon=True,
    fancybox=False,
    framealpha=0.92,
    facecolor="white",
    edgecolor="#D0D0D0",
    borderpad=0.35,
    handlelength=1.1,
    labelspacing=0.35,
    ncol=2,
)

plt.tight_layout()

plt.savefig("RQ2_impu.pdf", dpi=600, bbox_inches="tight", pad_inches=0.01)


plt.show()
```
